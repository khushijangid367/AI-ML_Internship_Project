```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#import data
Raw_mail_data  = pd.read_csv(r"C:\Users\khush\Downloads\mail_data.csv")
pd.set_option('display.max.rows',5575)
Raw_mail_data
```

```python
#data information
Raw_mail_data.info()
```

```python
#replace null values with ' ' space
mail_data = Raw_mail_data.where((pd.notnull(Raw_mail_data)),' ')
```

```python
#view top 5 values of data 
mail_data.head()
```

```python
#label spam mail as 0 and non spam mail as 1

mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

```python
#separate the data as text and labels 
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split data in train and test data
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 ,  random_state = 3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#feature extraction
#convert text into feature vectors 
feature_extraction = TfidfVectorizer(min_df =1 , stop_words = 'english', lowercase = True )
X_train_feature = feature_extraction.fit_transform(X_train)
X_test_feature = feature_extraction.transform(X_test)

#convert y test and y train into integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_feature)
```

```python
#training model
#X_train_feature LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train_feature , Y_train)
```

```python
#prediction on train data

prediction_on_training_data = model.predict(X_train_feature)
accuracy_on_training_data = accuracy_score(Y_train , prediction_on_training_data)
```

```python
print(accuracy_on_training_data)
```

```python
#prediction on test data

prediction_on_testing_data = model.predict(X_test_feature)
accuracy_on_testing_data = accuracy_score(Y_test , prediction_on_testing_data)
```

```python
print(accuracy_on_testing_data)
```

```python
#predictive model

input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]
#convert text to feature vectors
input_data_feature = feature_extraction.transform(input_mail)
#making prediction

prediction = model.predict(input_data_feature)
print(prediction)

if prediction[0]==1:
    print('non spam mail')
else:
    print('spam mail')
```
